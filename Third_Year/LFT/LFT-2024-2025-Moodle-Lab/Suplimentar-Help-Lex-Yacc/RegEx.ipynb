# Functions in the re module

### Match vs Search vs Find

```python
matches = re.match("c", "abcdef")
print(matches)
```

```python
matches = re.search("c", "abcdef")
print(matches)
```

```python
matches = re.findall("c", "abcdef")
print(matches)
```

```python
matches = re.finditer("c", "abcdef")
print(matches)

for match in matches:
  print(match)
```

### Restricting matches at the beginning of a string

```python
print(re.match("c", "abcdef"))
print(re.search("^c", "abcdef"))
print(re.search("^a", "abcdef"))
```

### Looping while pattern matching

```python
import time
```

```python
start = time.time()

for i in range(1000):
  re.match('hello', 'hello world')

end = time.time()
print(end - start)
```

```python
start = time.time()
h=re.compile('hello')

for i in range(1000):
  h.match('hello world')

end = time.time()
print(end - start)
```

```python
pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')

subbed_url = pattern.search(urls)
subbed_urls = pattern.findall(urls)

print(subbed_url)

for match in subbed_urls:
    print(match)
```
